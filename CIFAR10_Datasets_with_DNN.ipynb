```python
# Importing Basics Pacakges 
import numpy as np 
import pandas as pd 

import tensorflow as tf
from tensorflow import keras
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
print(x_train.shape)
print(x_test.shape )
print(y_train.shape)                    
print(y_test.shape )
```

```python
"""
Label	Description
0	airplane
1	automobile
2	bird
3	cat
4	deer
5	dog
6	frog
7	horse
8	ship
9	truck
"""
```

```python
class_name = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
class_name
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.imshow(x_train[8])
plt.colorbar()
plt.grid(True)
plt.show()
```

```python
# Data normalisation - min-max approach

x_train = x_train/255.0
x_test = x_test/255.0
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.imshow(x_train[5])
plt.colorbar()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize= (2,5))
for i in range (8):
    plt.subplot(2,4,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(True)
    plt.imshow(x_train[i], cmap = plt.cm.binary)
    plt.xlabel(class_name[y_train[i]])
plt.show()
```

```python
plt.figure(figsize= (10,10))
for i in range (10):
    plt.subplot(2,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(True)
    plt.imshow(x_train[i], cmap = plt.cm.binary)
    plt.xlabel(class_name[y_train[i]])
plt.show()
```

```python
tf.random.set_seed(100)
tf.keras.backend.clear_session()

model = None

model = keras.Sequential([keras.layers.Flatten(input_shape = (32,32,3)),
                        keras.layers.Dense(128, activation = 'relu'),
                        keras.layers.Dense(10)])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits= True),
             metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train, epochs = 10)
```
