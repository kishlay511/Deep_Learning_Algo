```python
# Import bsics package 
import numpy as np
import pandas as pd 
from sklearn.preprocessing import StandardScaler 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Baics Package - Deep Learning 
import tensorflow as tf
from tensorflow import keras
```

```python
from sklearn.datasets import load_boston
```

```python
X,y = load_boston(return_X_y = True)
```

```python
print(X.shape)
print(y.shape)
```

```python
X = pd.DataFrame(X)
```

```python
X.head()
```

```python
y = pd.DataFrame(y)
```

```python
y.head()
```

```python
# Feature Scaling 

sc = StandardScaler()
sc_x = sc.fit_transform(X)
sc_x
```

```python
X= pd.DataFrame(sc_x)
```

```python
X.head()
```

```python
# Split the data into train and test

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 101)
print (X_train.shape)
print (y_train.shape)
print (X_test.shape)
print (y_test.shape)
```

## Multiplayer Perceptron

```python
# Setup sequential API for building DNN

DNN = tf.keras.models.Sequential()
```

```python
# Input Layer connceted with 1st hidden layer

DNN.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))
```

```python
# Input Layer connceted with 2nd hidden layer

DNN.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))
```

```python
# Adding output layers 

DNN.add(tf.keras.layers.Dense(units=1))
```

```python
# Apply optimization techniques and measurement 

DNN.compile(optimizer= 'adam', loss= 'mean_squared_error')
```

```python
#Apply input dataset along with Backpropogation mehtod

DNN.fit(X_train, y_train, batch_size=32, epochs=500)
```

```python
# Predict the test data with traing models

y_pred = DNN.predict(X_test)
print(y_test)
print('#####'*25)
print(y_pred)
```
