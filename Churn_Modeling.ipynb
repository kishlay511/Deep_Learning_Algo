```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
dataset = pd.read_csv('C://Users//user/Desktop/Churn_Modelling.csv')
```

```python
dataset = dataset.iloc[:,3:]
```

```python
# Label Encoder
dataset['Geography'] = dataset['Geography'].astype('category')
dataset['Geography'] = dataset['Geography'].cat.codes
# One Hot Encoder
dataset = pd.get_dummies(dataset, columns = ['Geography'])
```

```python
# Label Encoder
dataset['Gender'] = dataset['Gender'].astype('category')
dataset['Gender'] = dataset['Gender'].cat.codes
# One Hot Encoder
dataset = pd.get_dummies(dataset, columns = ['Gender'])
```

```python
dataset.head()
```

```python
dataset = dataset.drop(['Geography_0', 'Gender_0'], axis = 1)
```

```python
dataset.head()
```

```python
x = dataset.drop(["Exited"], axis=1)
y = dataset['Exited']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_sc = sc.fit_transform(x)
x_sc
```

```python
x = pd.DataFrame(x_sc)
```

```python
# Imbalance data treatment 
import imblearn
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE()
x_smote, y_smote = smote.fit_resample(x,y)
print (x_smote.shape)
print (y_smote.shape)
```

```python
#Split the data into train and test
from sklearn.model_selection import train_test_split
x_train,  x_test,y_train, y_test = train_test_split(x_smote, y_smote, train_size = 0.75, random_state = 1520)
print (x_train.shape)
print (x_test.shape)
print (y_train.shape)
print (y_test.shape)
```

# Building Squential API

```python
# API
dnn = tf.keras.models.Sequential()
# Add first hidden layers
dnn.add(tf.keras.layers.Dense(units = 10, activation = 'relu'))
# Add second hidden layers
dnn.add(tf.keras.layers.Dense(units = 10, activation = 'relu'))
# Add third hidden layers
dnn.add(tf.keras.layers.Dense(units = 10, activation = 'relu'))
# Adding output layers
dnn.add(tf.keras.layers.Dense(units = 10, activation = 'relu'))
# Train the Deep Learning model (Multilayer Perceptron) for Classfication problem
dnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics= ['accuracy'])
# Apply the training dataset and use Backpropogation mehtod to fir the model
dnn.fit (x_train, y_train, batch_size = 32, epochs = 25)
```
